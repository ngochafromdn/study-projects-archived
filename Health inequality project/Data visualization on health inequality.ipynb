```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io
```

```python
uploaded = files.upload()
data_1 = pd.read_excel(io.BytesIO(uploaded["data_1.xlsx"]))
```

```python
data_1
```

```python
data_1.columns=["index","Country", "Code", "Inequality", "Current Health expenditure per capita","Population","Continent"]
```

```python
data_1
```

```python
from plotnine import *
```

```python
ggplot(data_1, aes(x = "Current Health expenditure per capita", y = "Inequality", color = "index")) + geom_point(size = 3)+ labs(y = "Inequality in life expectancy ", title = "Inequality in life expectancy vs. health expenditure per capital, 2015 ") + scale_color_brewer(type = "qual", palette = "Paired")
```
