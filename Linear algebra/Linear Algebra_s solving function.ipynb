## Reduce to row echelon form function


```python
def rowmul(matlist, rows, k,K):
    index=rows-1
    """
    Multiplies index row with k
    """
    for i in range(len(matlist[index])):
        matlist[index][i] = k*matlist[index][i]
    return matlist
```

```python
def rowadd(matlist, index1, index2, k,K):
    """
    Adds the index1 row with index2 row which in turn is multiplied by k
    """
    for i in range(len(matlist[index1])):
        matlist[index1][i] = (matlist[index1][i] + k*matlist[index2][i])
    return matlist
```

```python
def row_echelon(matlist, K):
    result_matlist = matlist
    nrow = len(result_matlist)
    for i in range(nrow):
        if (result_matlist[i][i] != 1 and result_matlist[i][i] != 0):
            rowmul(result_matlist, i, 1/result_matlist[i][i],K)
        for j in range(i + 1, nrow):
            if (result_matlist[j][i] != 0):
                rowadd(result_matlist, j, i, -result_matlist[j][i],K)
    return result_matlist
```

```python
row_echelon([[3, 7, 4],
             [2, 4, 5],
             [6, 2, 3]],2)
```

# Function using row operation to calculate determinant of a square matrix

```python

import copy
import sympy
def rowmul(matrix, index, k):
    for i in range(len(matrix[index])):
        matrix[index][i] = k*matrix[index][i]
    return matrix
def rowadd(matrix, index1, index2, k):
    for i in range(len(matrix[index1])):
        matrix[index1][i] = (matrix[index1][i] + k*matrix[index2][i])
    return matrix


def det_square(matrix):
    matrix2 = copy.deepcopy(matrix)
    rows = len(matrix2)
    det=1
    print(matrix2)
    print("____________________")
    for i in range(rows):
        if (matrix2[i][i] != 1 and matrix2[i][i] != 0):
            print("multiply row",i+1,"to",1/matrix2[i][i],"\n")
            det=det*matrix2[i][i]
            rowmul(matrix2, i, 1/matrix2[i][i])
            print(matrix2,"\n")
            print("det A=",det,'det',matrix2,"\n")
            print("____________________")

        for j in range(i + 1, rows):
            if (matrix2[j][i] != 0):
                print("add -(",matrix2[j][i],") row",i+1,"to row",j+1,"\n")
                rowadd(matrix2, j, i, -matrix2[j][i])
                print(matrix2, "\n")
                print("det A=",det,"det",matrix2, "\n")
                print("____________________")
    for i in range(rows):
      det=det*matrix2[i][i]
    return det
```

#Rotation clockwise function


```python
import math

def rotx(matrix,degree):
#rotation clockwise the positive x axis
  x=math.radians(degree)
  matrix2= sympy.Matrix([[1,0,0],[1,math.cos(x),-math.sin(x)],[0,math.sin(x),math.cos(x)]])
  print(matrix2)
  ma=matrix2*matrix
  return ma
def roty(matrix,degree):
#rotation clockwise the positive y axis
  x=math.radians(degree)
  matrix2= sympy.Matrix([[math.cos(x),0,math.sin(x)],[0,1,0],[-math.sin(x),0,math.cos(x)]])
  print(matrix2)
  ma=matrix2*matrix
  return ma
def rotz(matrix,degree):
#rotation clockwise the positive y axis
  x=math.radians(degree)
  matrix2= sympy.Matrix([[math.cos(x),-math.sin(x),0],[math.sin(x),math.cos(x),0],[0,0,1]])
  print(matrix2)
  ma=matrix2*matrix
  return ma


```
